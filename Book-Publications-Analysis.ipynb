```python
Place_of_Publication = df_books['Place of Publication']
london = Place_of_Publication.str.contains('London')
oxford = Place_of_Publication.str.contains('Oxford')
```

```python
df_books['Place of Publication'] = np.where(london, 'London', 
                                            np.where(oxford, 'Oxford', 
                                                     Place_of_Publication.str.replace('-',' ')))
df_books['Place of Publication'].head()
```

```python
# Convert the 'Date of Publication' column to numeric, forcing errors to NaN
df_books['Date of Publication'] = pd.to_numeric(df_books['Date of Publication'], errors='coerce')

df_london = df_books[(df_books['Place of Publication'] == 'London') & (df_books['Date of Publication'] >= 1800)]
df_paris = df_books[(df_books['Place of Publication'] == 'Paris') & (df_books['Date of Publication'] >= 1800)]
df_edinburgh = df_books[(df_books['Place of Publication'] == 'Edinburgh') & (df_books['Date of Publication'] >= 1800)]
df_new_york = df_books[(df_books['Place of Publication'] == 'New York') & (df_books['Date of Publication'] >= 1800)]
df_leipzig = df_books[(df_books['Place of Publication'] == 'Leipzig') & (df_books['Date of Publication'] >= 1800)]


books_london = df_london['Date of Publication'].value_counts().sort_index()
books_paris = df_paris['Date of Publication'].value_counts().sort_index()
books_edinburgh = df_edinburgh['Date of Publication'].value_counts().sort_index()
books_new_york = df_new_york['Date of Publication'].value_counts().sort_index()
books_leipzig = df_leipzig['Date of Publication'].value_counts().sort_index()

plt.figure(figsize=(20,8))

plt.plot(books_london.index, books_london.values, label='London')
plt.plot(books_paris.index, books_paris.values, label='Paris')
plt.plot(books_edinburgh.index, books_edinburgh.values, label='Edinburgh')
plt.plot(books_new_york.index, books_new_york.values, label='New York')
plt.plot(books_leipzig.index, books_leipzig.values, label='Leipzig')

plt.tick_params(axis='x', labelrotation=45)
plt.title('Book Publication Trends in Major Cities(1800-1900)')
plt.xlabel('Publication Year', fontsize=15)
plt.ylabel('Total Publications', fontsize=15)
plt.legend(bbox_to_anchor = (1,.50))

plt.savefig('C:/Users/murta/Downloads/Book-Publications-Trend-Lineplot.pdf', dpi=128, bbox_inches='tight')

plt.show()
```
